Exercise 3
Use of Persona Prompt, Root Prompt, Cognitive Verifier Pattern, Question Refinement Pattern, Provide New Information and Ask Questions

1.Root Prompt
A Root Prompt defines the base behavior of the AI:

How it should think

How it should explain

Tone and style

It acts like instructions given before starting a conversation.

```python
print("\n--- ROOT PROMPT ---\n")

root_prompt = """
You are an expert Generative AI tutor.
Explain concepts clearly, step by step, using simple language.
"""

print(ask_ai(root_prompt + "Explain Prompt Engineering."))
```

2.Persona Prompt
A Persona Prompt assigns a role or identity to the AI (teacher, doctor, mentor, etc.).
This changes how the AI explains, not what it explains.

```python
print("\n--- PERSONA PROMPT ---\n")

persona_prompt = """
Act as a college professor teaching machine learning to undergraduate students.
Explain Prompt Engineering with real-life examples.
"""

print(ask_ai(persona_prompt))
```

3.Cognitive Verifier Pattern
Here the AI:

Answers the question

Checks its own answer

Mentions missing or weak points

This improves accuracy and reliability.

```python
print("\n--- COGNITIVE VERIFIER PATTERN ---\n")

cognitive_prompt = """
Explain what Prompt Engineering is.
After answering, verify your explanation and mention if anything is missing.
"""

print(ask_ai(cognitive_prompt))
```

4. Question Refinement Pattern
Used when a question is too vague or unclear.
AI rewrites it into a better, more precise question.

```python
print("\n--- QUESTION REFINEMENT PATTERN ---\n")

refine_prompt = """
Refine the following question to make it clearer:
"What is AI?"
"""

print(ask_ai(refine_prompt))
```

Provide New Information and Ask Questions

The AI:

Explains the topic

Adds extra related knowledge

Asks follow-up questions to continue learning

Very useful for interactive learning.

```python
print("\n--- PROVIDE NEW INFO & ASK QUESTIONS ---\n")

info_prompt = """
Explain Machine Learning.
Provide new related information.
Then ask two follow-up questions.
"""

print(ask_ai(info_prompt))
```

Exercise 4
Chain of Thought, Tabular Format, Fill in the Blank, RGC, Zero-shot, One-shot, Few-shot Prompting

Chain of Thought Prompting


AI explains its reasoning step by step, instead of jumping to the answer.
Helps in understanding how decisions are made.

```python
print("\n--- CHAIN OF THOUGHT ---\n")

cot_prompt = """
Explain step by step how AI detects spam emails.
Show your reasoning clearly.
"""

print(ask_ai(cot_prompt))
```

Tabular Format Prompting
Used when information needs comparison or structure.
AI outputs data in table form.

```python
print("\n--- TABULAR FORMAT ---\n")

table_prompt = """
Explain the difference between AI, ML, and Deep Learning in a table.
"""

print(ask_ai(table_prompt))
```

Fill in the Blank Prompting
AI completes missing information.
Commonly used in education and assessments.

```python
print("\n--- FILL IN THE BLANK ---\n")

blank_prompt = """
Fill in the blank:
Artificial Intelligence is the ability of machines to ______ human intelligence.
"""

print(ask_ai(blank_prompt))
```

RGC (Role–Goal–Context) Prompting

Prompt is structured using:

Role – Who AI is

Goal – What AI should do

Context – Situation of the user

Produces highly relevant answers.

```python
print("\n--- RGC PROMPTING ---\n")

rgc_prompt = """
Role: You are an AI mentor.
Goal: Explain prompt engineering.
Context: The learner is a beginner with no AI background.
"""

print(ask_ai(rgc_prompt))
```

Zero-Shot Prompting

No examples are provided.
AI answers purely based on its training.

```python
print("\n--- ZERO SHOT PROMPTING ---\n")

zero_shot = "Explain Natural Language Processing."
print(ask_ai(zero_shot))
```

One-Shot Prompting

Explanation

One example is given so AI follows the same pattern.

```python
print("\n--- ONE SHOT PROMPTING ---\n")

one_shot = """
Example:
Input: What is AI?
Output: AI is the simulation of human intelligence by machines.

Now explain:
What is Machine Learning?
"""

print(ask_ai(one_shot))
```

Few-Shot Prompting

Explanation

Multiple examples are provided.
AI learns the pattern and completes the task.

```python
print("\n--- FEW SHOT PROMPTING ---\n")

few_shot = """
Input: AI
Output: Machines that simulate human intelligence.

Input: ML
Output: Systems that learn from data.

Input: Deep Learning
Output:
"""

print(ask_ai(few_shot))
```
